```python
teams = t4["gname"].unique()
decision_making = pd.DataFrame([],columns= ['Gname','success','Times'])
for evenid, element in enumerate(teams):
    group0 = t4[(t4['gname']== element) & (t4['success']== 0)]
    group1 = t4[(t4['gname']== element) & (t4['success']== 1)]
    
    decision_making = decision_making.append({'Gname': element,'success': 0,'Times' : group0['gname'].count()}, ignore_index= True)
    decision_making = decision_making.append({'Gname': element,'success': 1,'Times' : group1['gname'].count()}, ignore_index= True)
```

```python
# creating histograms to visualize all the data
fig = plt.figure(figsize = (40,40))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

ax = fig.gca()
t4.hist(ax = ax)
```

```python
ps = t4.groupby('attacktype1_txt')[['suicide','success']].sum().sort_values(by='success',ascending=False)
ps[:].plot.bar(color=['blue','green'],figsize=(15,25))
plt.title('attacktype')
plt.xlabel('attacktype1_txt')
plt.ylabel('suicide/success')
plt.show()
```

Suicide bombing and explosion is use more for attacking the people

Bombing and Explosion is use more common by terror group

```python
plt.figure(figsize=(15,15))
sns.countplot(x=t4['region_txt'])
plt.xticks(rotation=90)
plt.title("most affected area")
plt.show()
```

Middle East and North Africa most affected by terror groups

```python
plt.figure(figsize=(15,15))
sns.countplot(x=t4['weaptype1_txt'])
plt.xticks(rotation=90)
plt.title("weapon use by terror groups")
plt.show()
```

Above graph shows that terror groups use most number of time explosives to make huge destruction

```python
plt.figure(2, figsize=(20,15))
sns.barplot(x=t4['region_txt'],
           y=t4['success'].values,
           data = t4)
plt.xticks(rotation= 70)
plt.title('region_txt/success')
plt.xlabel('region_txt')
plt.ylabel('success')
plt.show()
```

```python
plt.figure(2, figsize=(20,15))
sns.barplot(x=t4['region_txt'],
           y=t4['suicide'].values,
           data = t4)
plt.xticks(rotation= 70)
plt.title('Categoryregion_txt/suicide')
plt.xlabel('region_txt')
plt.ylabel('suicide')
plt.show()
```
